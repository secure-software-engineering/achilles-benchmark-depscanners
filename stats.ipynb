```python
import pandas
import os
import matplotlib.pyplot as plt
import json
```

```python
import fnmatch
import io

def genDataframeFor(filenamePattern="*.json"):
    matches = []

    for root, dirnames, filenames in os.walk('./detection'):
        for filename in fnmatch.filter(filenames, filenamePattern):
            matches.append(os.path.join(root, filename))

    li = []

    for filename in matches:
        with open(filename) as f:
            df = json.load(f)
            li.append(df)
    frame = pandas.DataFrame(li)
    return frame
```

```python
df = genDataframeFor()
```

```python
df
```

```python
gavs = pandas.json_normalize(df['gav'])
gavs
```
